DILASI


import library


```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

import gambar


```python
citra = cv2.imread('image/koin.jpeg')
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
dilasi = cv2.dilate(citra, kernel)
```

Tampilkan citra asal dan hasil


```python
plt.subplot(1,2,1)
plt.imshow(citra)
plt.title("Gambar asli")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(dilasi)
plt.title("Dilasi")
plt.axis('off')

plt.show()
```

EROSI


```python
citra = cv2.imread('image/koin.jpeg')
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
erosi = cv2.erode(dilasi, kernel)
```

Tampilkan hasil erosi


```python
plt.subplot(1,2,1)
plt.imshow(dilasi)
plt.title("Dilasi")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(erosi)
plt.title("Erosi")
plt.axis('off')

plt.show()
```

Erosi dengan iterasi


```python
citra = cv2.imread('image/koin.jpeg')
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
dilasi = cv2.dilate(citra, kernel, iterations=3)
erosi = cv2.erode(dilasi, kernel, iterations=3)
```

```python
plt.subplot(1,2,1)
plt.imshow(dilasi)
plt.title("Dilasi 3 iterasi")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(erosi)
plt.title("Erosi 3 iterasi")
plt.axis('off')

plt.show()
```

Opening


```python
citra = cv2.imread('image/koin.jpeg')
kernelA = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3,3))
kernelB = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
erosi = cv2.erode(citra, kernelA)
openingA = cv2.morphologyEx(citra, cv2.MORPH_OPEN, kernelA)
openingB = cv2.morphologyEx(citra, cv2.MORPH_OPEN, kernelB)
baris1 = np.hstack((citra, erosi))
baris2 = np.hstack((openingA, openingB))
hasil = np.vstack((baris1, baris2))
```

```python
plt.subplot(1,2,1)
plt. imshow(citra)
plt.title("Gambar asli")
plt.axis('off')

plt.subplot(1,2,2)
plt. imshow(erosi)
plt.title("Erosi")
plt.axis('off')

plt.show()
```

```python
plt.subplot(1,2,1)
plt. imshow(openingA)
plt.title("Opening A")
plt.axis('off')

plt.subplot(1,2,2)
plt. imshow(openingB)
plt.title("Opening B")
plt.axis('off')

plt.show()
```

```python
plt.imshow(hasil)
plt.title('Hasil')
plt.axis('off')
plt.show()
```

Closing


```python
citra = cv2.imread('image/koin.jpeg')

# konversi ke binner
ambang, citra = cv2.threshold(citra, 180, 255, cv2.THRESH_BINARY)

# Balik warna
citra = 255 - citra
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (7,7))
closing = cv2.morphologyEx(citra, cv2.MORPH_CLOSE, kernel)

# tampilkan citra dan hasil

plt.subplot(1,2,1)
plt.imshow(citra)
plt.title('citra asal')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(closing)
plt.title('closing')
plt.axis('off')

plt.show()
```
